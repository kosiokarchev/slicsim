```python
import numpy as np
import pandas as pd
import seaborn as sns
import torch
from matplotlib import pyplot as plt
from scipy.stats import norm

from clipppy.utils.plotting.nre import to_percentiles


sns.reset_orig()


x = np.linspace(0, 1, 1001)
b = 0.42
p = (1-b) * norm.pdf(x, loc=0.7, scale=0.08) + b * norm.pdf(x, loc=0.42, scale=0.26)
perc = to_percentiles(torch.tensor(p)).numpy()

plt.plot(x, p)
plt.plot(x, perc)
plt.axhline(0.68, color='k')
plt.xlim(0, 1)
plt.ylim(bottom=0)

pd.DataFrame({'x': x, 'p': p, 'perc': perc}).to_csv('paper/plotdata/freqcal.csv', index=False)
```
